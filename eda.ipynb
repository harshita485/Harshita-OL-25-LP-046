```python
st_clean = st[(st['Age'] >= 18) & (st['Age'] <= 80)]
print("\n Cleaned age data:")
print(st_clean['Age'].describe())
```

```python
from typing import Iterable
def clean_employees_column(series:Iterable[object]) -> Iterable[object]:
    """ Function to deal with no_employees column 
    parameters
    series: Iterable[object], usually a string as per the dataset
     returns series"""
    cleaned=[]
    for val in series:
        val_str = str(val).strip().lower()
        if '-' in val_str:
            try:
                parts= val_str.split('-')
                nums= [int(p.strip()) for p in parts if p.strip().isdigit()]
                if len(nums) == 2:
                    cleaned.append(sum(nums)//2)
                else:
                    cleaned.append(np.nan)
            except:
                cleaned.append(np.nan)

        elif 'more than' in val_str:
            tokens = val_str.split()
            try:
                num_index = tokens.index('than') + 1
                number = int(tokens[num_index])
                cleaned.append(number+1)
            except:
                cleaned.append(np.nan)
        elif val_str.isdigit():
            cleaned.append(int(val_str))

        else:
            cleaned.append(np.nan)
    cleaned_series = pd.Series(cleaned)

   #Replace NaNs with mean of valid entries
    mean_value=cleaned_series.dropna().mean()
    cleaned_series.fillna(mean_value, inplace=True)
    return cleaned_series.astype(int)
st_clean['cleaned_employees'] = clean_employees_column(st_clean['no_employees'])


        

            

    
```

```python
st_missing = st_clean.copy()
missing_treatment_strategy={
    'state': 'unknown',
    'self_employed': 'mode',
    'work_interfere': 'mode',
    'comments' : 'unknown'

}
for column, strategy in missing_treatment_strategy.items():
    if column in st_missing.columns:
        missing_count= st_missing[column].isnull().sum()
        missing_pct = (missing_count / len(st_missing)) * 100
        print(f"\n{column}: {missing_count} missing ({missing_pct:.2f}%)")
        if strategy == 'mode' and missing_count > 0:
            mode_value = st_missing[column].mode()[0] if not st_missing[column].mode().empty else 'unknown'
            st_missing[column].fillna(mode_value, inplace=True)
            print(f"Filled missing values in '{column}' with mode: {mode_value}")
        elif strategy == 'unknown' and missing_count > 0:
            st_missing[column].fillna('unknown', inplace=True)
            print(f"Filled missing values in '{column}' with 'unknown'")
print(f"\n Dataset shape after missing value treatment: {st_missing.shape}")
print(f"Missing values after treatment:\n{st_missing.isna().sum()}\n")
```

```python
def clean_gender(g):
    if pd.isna(g):
        return 'Not specified'
    g=str(g).lower().strip()
    if any(term in g for term in['male', 'm','man','boy','guy']):
        if not any(term in g for term in['female','woman', 'girl', ] ):
            return 'Male'
    if any (term in g for term in ['female', 'f', 'woman', 'girl', 'lady']):
        return 'Female'
    if any(term in g for term in ['non-binary', 'nb', 'queer', 'agender', 'genderfluid', 'trans']):
        return 'Other'
    return 'Not specified'
   
```

```python
st_missing['Gender_clean'] = st_missing['Gender'].apply(clean_gender)
print(st_missing['Gender_clean'].unique())
st_missing['Gender_clean'].value_counts(dropna=False)
```

## Dealing with categorical data 

```python
categorical_columns=['treatment','work_interfere','re,ote_work','tech_company',
                   'benefits' ,'care_options','wellness_program','seek_help',
                    'anonymity','leave','mental_health_consequence',
                     'phys_health_consequence','cowworkers','supervisor','mental_vs_physical' ]
#REPLACING VARIOUS FORMS OF MISSING/UNKNOWN VALUES
replace_dict={
    'Don\'t know': 'unknown',
    'don\'t know': 'unknown',
    'Not sure': 'unknown',
    'not sure': 'unknown',
    'Maybe': 'unceratin',
    'maybe': 'unceratin' ,
}
for col in categorical_columns:
    if col in st_clean.columns:
        st_missing[col] = st_missing[col].replace(replace_dict)

print("data cleaning completed")
print(f"Final dataset shape: {st_missing.shape}")
```

# now dealing with  numerical outliers 

```python
numerical_columns=st_missing.select_dtypes(include=[np.number]).columns
print(f"Numeical columns for outlier analysis:{list(numerical_columns)}")

outlier_results ={}

for col in numerical_columns:
    if col in st_missing.columns and st_missing[col].notna().sum() > 0:
        data= st_missing[col].dropna()
        q1 = data.quantile(0.25)
        q3 = data.quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = data[(data < lower_bound) | (data > upper_bound)]
        outlier_count = len(outliers)
        outlier_pct = (outlier_count / len(data)) * 100
        outlier_results[col] = {
            'total_count': len(data),
            'outlier_count': outlier_count,
            'outlier_pct': outlier_pct,
            'Bounds': (lower_bound, upper_bound),

            'Outliers': outliers.tolist()
        }
print("\nOutlier Detection Summary")
for col, results in outlier_results.items():
    print(f"\nColumn: {col}")
    print(f"Total values: {results['total_count']}")
    print(f"Outlier Count: {results['outlier_count']} ({results['outlier_pct']:.1f}%)")
    print(f"Bounds: {results['Bounds'][0]:.2f} to {results['Bounds'][1]:.1f}")
    if results['Outliers']:
        print(f"Outliers: {results['Outliers'][:5]}...")
lower_bound, upper_bound = results['Bounds']
outlier_mask = (st_missing[col] < lower_bound) | (st_missing[col] > upper_bound)
st_outlier = st_missing[~outlier_mask]
print(f"\n Dataset shape sfter outlier removal: {st_outlier.shape}")
print(f"Original dataset shape: {st_missing.shape}")
print(f"Rows removed due to outliers: {st_missing.shape[0] - st_outlier.shape[0]}")
st_clean=st_outlier.copy()
    
        
```

# Multivariate Analyisis

```python
st_correlation = st_clean.copy()
# binary encoding implemenation for categorical variables
binary_mapping={
    'treatment': {'Yes': 1, 'No': 0},
    'family_history': {'Yes': 1, 'No': 0},
    'remote_work': {'Yes': 1, 'No': 0},
    'tech_company': {'Yes': 1, 'No': 0},
    'benefits': {'Yes': 1, 'No':0, 'Unknown':0.5},
    'care_options': {'Yes': 1, 'No': 0,'Unknown':0.5},
    'wellness_program': {'Yes': 1, 'No': 0, 'Unknown': 0.5},
    'anonymity': {'Yes': 1, 'No': 0, 'Unknown': 0.5},
    'leave':{ 'Very easy': 1, 'Somewhat easy': 0.75, 'Somewhat difficult': 0.25, 'Very difficult': 0, 'Unknown': 0.5},
    'mental_health_consequence': {'Yes': 1, 'No': 0, 'Maybe': 0.5, 'Unknown':0.5},
    'phys_health_consequence': {'Yes': 1, 'No': 0, 'Maybe': 0.5, 'Unknown':0.5}

}
# binary mapping 
for col, mapping in binary_mapping.items():
    if col in st_correlation.columns:
        st_correlation[col +'_binary'] = st_correlation[col].map(mapping)

numerical_columns=['Age', 'treatment_binary', 'family_history_binary', 'remote_work_binary',
                   'tech_company_binary', 'benefits_binary', 'care_options_binary',
                   'wellness_program_binary', 'anonymity_binary', 'leave_binary',
                   'mental_health_consequence_binary', 'phys_health_consequence_binary']
```

```python
corr_data= st_correlation[numerical_columns].dropna()
correlation_matrix = corr_data.corr()
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix,annot=True, cmap='coolwarm', center=0, square=True, cbar_kws={"shrink" : .8}, linewidths=0.5)
plt.title("Correlation Matrix  of Mental Health survey data", fontweight='bold', fontsize =16)
plt.tight_layout()
plt.show()
# strong corelations with treatment
treatment_corr= correlation_matrix['treatment_binary'].sort_values(ascending=False)
print(treatment_corr[abs(treatment_corr) > 0.1])
```
